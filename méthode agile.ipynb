# Méthode Agile Scrum

### Qu’est ce que la méthode Scrum ? Quel est son objectif ?

La méthode agile est une méthode de gestion des projets dévéloppée par 17 experts en développement de logiciel en 2001 qui forme une alternative à la méthode de "Cycle en V" ou la méthode de "Cycle en cascade".

Suite à l'observation d'un taux d'échec des projets assez élevé, ils décident de régrouper leurs différentes méthodes de travail dans le "Manifeste Agile" qui regroupe alors les valeurs et les principes fondamentaux de la méthode. 

Les principes de base de la méthode agiles sont: 
    * Une planification des objectifs à court terme, le projet est ainsi divisé en sous projet => plus de flexibilité.     
    * La satisfaction client est une priorité => + d'implication du client + de communication.
    
La méthode agile contient plusieurs démarches qui gardent le même objectif et différent par leurs outils et quelques nuances organisationnelles. La plus répandu est la méthode **agile Scrun**.

ELle s'organise autour des cycles dit "sprint" avec des rôles clés, des rituels et des événements clé.

=> Scrun est donc une approche de gestion de projet dynamique, participative et basée sur l'expérience 



### La méthode Scrum repose sur des valeurs, des principes et des piliers. Que sont-ils ? Expliquez les.

**- Les valeurs:** F.O.R.C.E: L’objectif de ces valeurs est d’aider les équipes à créer et renforcer un mindset agile

* Focus: Il n’y a pas de place pour le multitâches. L’équipe Scrum doit finir ce qu’elle a commencé avant d’entamer un nouveau travail -> l'équipe porte son attention sur le produit qu’elle réalise, ses objectifs à atteindre et les tâches en cours de réalisation.

* Ouverture: Ouverture aux nouvelles idées, aux nouvelles manières de faire, Ouverture d'esprit -> Suivre les changement. 

* Respect: Respect des parties prenantes, des temps de paroles,.... -> Faire reigner un climat de confiance. 

* Courage: Avoir le courage de faire face aux challenges et aux problèmes liés au projet, de prendre des risques, de tester des nouvelles solutions (des fois pas très populaire dans l'équipe), avoir le courage de dire "NON". 

* Engagement: respecter les échéances, fournir le meilleur de soi même, apprendre et s'améliorer continuellement. 


**-Principes:**

* Processus empirique: agir et prendre des décisions sur la base de faits, d’expériences, d’observations et de preuves


* Auto-organisation: - disposer d’un certain niveau de pouvoir décisionnel
- prendre possession de sa façon de travailler 
- pas de compte à rendre à un chef de projet. le rôle du SM:conseiller l’équipe.
- expliquer au SM et PO son organisation
                     
* Collaboration: maintenir une communication avec les parties prenantes pour prendre en compte les remarques et les retours d'expérience.

* Priorisation basée sur la valeur: organiser les tâches à réaliser en fonction de leur importance et leurs valeurs apportée au projet d'un point de vue client. 

* Délimité dans le temps: pour maintenir un niveau d'efficience de l'équipe, et éviter les réunions interminables. 

* Développement itératif: dévélopper des fonctionnalités plutôt que développer un produit entier d'un coup. => accélérer le rythme de production, vérifier que le résultat répond aux attentes finales et apporter les améliorations nécessaires. 


**-Les piliers:**

* La transparence: Présenter les faits tels qu'ils sont, sans déformation nu minimisation 
Les objectifs du projet, l'attendu et la progression du projetw<QS sont partagés, communiqués par les utilisateurs finaux.
                   
* l'inspection: elle est réalisée par tout les membres de l'équipe et concerne aussi bien les qualités du livrable que les process, les méthodes de travail, les pratiques de l'équipe. l'objectif de l'inspection est l'amélioration continue grâce à une prise de recul permanente. 

* l'adaptation: Elle concerne l'évolution par rapport aux besoins du client, l'évolution du marché et des techniques utilisées. 

### Toute équipe de méthodes agiles est auto organisée. Comment se compose une équipe Scrum ? Expliquez leur rôle et leur importance.

L'équipe Scrum est compsée de:
* **l'équipe de dévéloppeurs:** Elle regroupe toutes les personnes qui réalisent le travail. Elle n'est pas composée exclusivement d'ingénieurs mais aussi des designers, des rédacteurs ou d'autres métiers intervenant dans la création du livrable. 
Elle s'auto gére et peut prendre des décisions dans la réalisation du travail.
Ces responsabilités concernent: 
    1. la livraison du travail
    2. La participation à réunion quotidienne afin de garder la tranparence nécessaire au bon déroulent du travail. 
    
    
    
* **Product Owner:** Son rôle principal est de s'assurer que son équipe offre la meilleur valeur possible. Ainsi, il représente son entreprise auprès du client. Ce qui lui demande un équilibre entre le besoin du client, la vision de l'équipe Scrum et les différentes parties prenantes au projet. Sa mission est alors de hiérarchiser le travail en conjugant les besoins de chacun. 
Ses responsabilités sont: 
    1. La gestion du Product Backlog: Il est responsable de toutes les tâches inscrites au product backlog. 
    2. La gestion des livraisons: Il planifie les livraisons en fonction du besoin du client et des moyens de l'équipe de dévéloppeurs.
    3. La gestion des parties prenantes: Il collabore avec toutes les personnes impliqués dans le projet afin de s'assurer que l'équipe de dévéloppeurs apporte une valeur ajoutée.
    
    

* **Le Scrum Master:** Il aide le PO à définir la valeur à fournir par l'équipe de dévéloppeurs (User Stories), à comprendre les besoins des parties prenantes et à gérer le product backlog. Il veille à la bonne application de la méthode Scrum. Il joue donc un rôle de leader serviteur et assume la responsabilité des bloqueurs qui n'entrent pas dans la capacité de l'équipe de dévéloppeurs. Il s'assure que chaque opportunité d'amélioration est transparente pour l'équipe Scrum.

### Un projet organisé par la méthode Scrum est rythmé par un ensemble d'événements. Énoncez ces différents événements et expliquez les.

**1. Le Sprint Planning:** 

Il résume l'objectif du sprint (Sprint Goal) ainsi que les fonctionnalités à livrer du produit. 

Seul le PO peut invalider le sprint goal (C'est pourtant très rare). 

Toute l'équipe Scrum définit le Sprint Goal et les Users Stories à réaliser pendant le sprint. 

L'équipe des dévéloppeurs s'engage sur le sprint goal et les Users Stories à réaliser. Elles sont donc inscrites dans le Sprint Backlog. 

Le Sprint Planning dure maximun 8H pour un sprint de 1 mois.


**2. Le planning Poker:**
Durant le Planning Poker, l'équipe de dévéloppeur, en présence du Poduct Owner, propose un score aux différentes User Stories en fonction de leurs difficultés. 
Ce système de scoring, permet d'estimer les ressources nécessaires à la réalisation de chaque user story.


**3. Le Daily Scrum: La mêlée quotidienne:**
Elle dure 15 minutes tout les jours et permet la communication entre l'équipe Scrum. C'est l'occasion à chaque membre de se sychroniser avec les reste de l'équipe et de demander de l'aide si besoin. Pour plus d'efficacité, les participants répondent à 3 questions: 
    * qu'est ce que j'ai fait hier?
    * qu'est ce que vais faire aujourd'hui?
    * qu'elles sont les difficultés rencontrées? 
    
    
**4. La Sprint Review:** 
Toutes les parties prenantes assistent à cette réunion afin de faire le point sur ce qui a été réalisé durant le Sprint et de voir si l’objectif du Sprint est atteint.
Durant la Sprint Review l'équipe Scrum présente les users stories rencontrées, répond aux questions des participants et prend note des retours. 
Elle permet au PO de faire le point sur le product backlog et d'y apporter des ajustements. 
ELle dure maximun 4H. 


**5. La rétrospective:**
A la fin du sprint, c'est l'occasion de se focaliser sur les personnes, analyser les relations et les process qui ont permis la réalisation (ou pas) des users stories. 
ELle peut être illustrée par plusieurs outils mais son déroulé suit les étapes suivantes: 
    * La mise en place
    * La récolte des données
    * La génération des idées
    * La décision des actions à réaliser
    * Cloturer la rétrospective
    
Elle dure 3h maximum

### Donnez les artefacts et outils de la méthode Scrum. Les artefacts désignent des produits ayant subi des transformations durant la gestion du projet.

**1. Le product Backlog**: 

C'est un document évolutif (enrichi en permanence) qui regroupe tous les besoins et fonctionnalités demandées par le client, l'utilisateur ou l'équipe Scrum. 

Il est composé des User Stories qui définissent le besoin à satisfaire par les fonctionnalités demandées. => L'objectif du Product Backlog est de faciliter la compréhension du besoin et minimiser les quiproquos.

Le Product Owner est responsable de son organisation et de la priorisation des users stories

Le Product Backlog répond au Product Goal: Pourquoi ce projet? Quel est le but de ce projet? 

Le Product Goal peut être amené à évoluer durant le projet en fonctions des feedbacks des parties prenantes, le Product Backlog évolue également. 

Si le Product Goal est atteint, soit le dévéloppement du produit prend fin, soit un nouvel objectif est fixé. 


**2.Le Sprint Backlog:**

Il contient les User Stories et les beugs à corriger durant le sprint. 

La définition du Sprint Backlog a lieu durant le Sprint Planning. 

Durant le Sprint, les Users Stories peuvent être décomposées en différentes tâches, une fois que les dévéloppeurs ont plus de maitrise du sujet. 

Toute l'équipe de dévéloppeurs est responsable du Sprint Backlog (le mettre à jour, piocher une tâche à réaliser). 

Le Sprint Backlog précise le Sprint Goal, soit l'incrément de valeur à produire durant la sprint. Il doit être clair pour: 
        - Motiver les équipes 
        - Se focaliser sur les tâches les plus importantes. 
        - Décider des différentes options
        - Gérer les risques
        - Partager une cible à court terme pour toutes les parties prenantes. 


**3.L'incrément:**

C'est le livrable partiel de la fin de chaque sprint et il constitue un pas de plus vers le produit final. Il s'agit d'un prototype, il est donc opérationnel à l'issue du premier sprint.

Il est ajusté durant les prochaines itérations pour réponde au mieux aux attentes du client. 

/!\ il est important de définir de manière unanime le terme "terminé" en précisant une liste de vérifications des tâches que l'équipe doit réaliser avec succès pour déclarer que la fonctionnalité est livrable. 

### Définissez de manière claire et concise, les mots clés suivants : user story, story points, la DoD, un Burndown chart et la méthode Kanban.

*1. User Story:* 

Les user stories désignent une série de phrases, rédigées dans un langage simple, qui décrivent le résultat souhaité. Elles n'entrent pas dans le détail. Des exigences sont ajoutées par la suite, une fois convenues par l'équipe.

Lors de la rédaction des User Stories, il est important de tenir compte des points suivants:

    -> Définir quand est ce qu'une US est considérée comme terminée. 
    -> Définir les tâches ou les sous tâches
    -> Définir le Personas utilisateur
    -> Organiser les stories pour chaque étape
    -> Tenir compte des feedbacks 
    -> Donner une idée sur la durée allouée à une US

Une user story prend alors la forme: « En tant que [persona], je [souhaite que] [afin de] ».

*2. Story Points:* 

Il s'agit d'un numéro, ou valeur, est attribué à chaque user story (ou scénario utilisateur) afin d'évaluer l'effort total (énergie, charge de travail, etc.) à allouer à sa réalisation. Par conséquent, l’estimation en story points doit considérer tous les paramètres amenés à impacter cet effort :

    -  quantité de travail à fournir.
    -  complexité des tâches à exécuter.
    - Tout risque ou incertitude qui viendraient compromettre leur réalisation.
    - les capacités techniques de l'équipe.
    
*3. DoD:*

C'est une liste de vérifications des tâches que l’équipe doit réaliser avec succès avant de déclarer que la fonctionnalité est potentiellement livrable.
ELle doit être entendue, comprise et appliquée par toute l’équipe. 

*4. Burndown Chart:* 

C’est un indicateur de mesure qui montre le travail restant sur une période donnée. Il permet de suivre le projet dans le temps afin de contrôler sa livraison dans les délais souhaités. 




*5. Kanban :* 

Il s'agit d'une approche Agile axée sur la visualistion du travail afin de réduire le travail en cours et gagner en efficacité. Elle utilise le tableau Kanban pour illustrer l'avancement de tâches: "à réaliser", "en cours", "finis". 
Ce tableau est également utilisée dans l'approche Agile Scrum. 

### Étant très impressionné par la puissance et l’efficacité de la méthode Scrum, vous décidez d’argumenter son utilisation aux autres étudiants.

Suite à la réalisation d'un projet en méthode agile, j'ai pu prendre conscience de la puissance et l'efficacité de cette méthode. En effet, grâce aux rôles déjà définis, les responsabilités de chaque membre de l'équipe étaient claires. Le fait que l'équipe de dévéloppeur était auto-gérée, a permis la responsabilisation et la montée de compétence de chaque membre de l'équipe, ce qui a généré de l'auto satisfaction et amélioré l'implication de membres de l'équipe. 

Egalement, les cérémonies sont bien articulées ce qui limite les réunions interminables et garantie le temps de parole de chaque membre de l'équipe tout en restant informé de l'évolution des différents aspect du projet. 

La retrospective était un moment clé du sprint car elle permet d'éclaircir les points d'ombre qui ont pu se créer durant le sprint et ainsi prendre le recul nécessaire pour s'améliorer de sprint en sprint.

Une adepte de la communication, les cérémonies sont pour moi un moment clé pour échanger, apprendre et exprimer son point de vue et écouter les avis des autres membres.
